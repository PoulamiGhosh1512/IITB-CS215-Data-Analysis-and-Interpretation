### Laplace distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
def Laplace(x,mu,b):
    return ((1/(2*b))*exp(-abs(x-mu)/b))
```

```python
xlist = np.linspace(-15,15,num=1000)
```

```python
ylist = [Laplace(x,mu=2,b=2) for x in xlist]
```

```python
plt.figure(num=0,dpi=120)
plt.plot(xlist,ylist)
plt.title('Laplace Distribution')
plt.xlabel('x')
plt.ylabel('PDF')
plt.savefig('Laplace Distribution PDF')
```

```python
def RiemannSum_Laplace(a, b, N): #Computing Left Riemann Sum
    cdf = []
    delta_x = (b-a)/N 
    Csum = 0.0
    x = a     
    for i in range(N):
        Csum += Laplace(x,mu=2,b=2)*delta_x
        cdf.append(Csum)
        x = x + delta_x 
    return cdf
```

```python
CDF_Laplace = RiemannSum_Laplace(-15,15,1000)
```

```python
plt.figure(num=0,dpi=120)
plt.plot(xlist,CDF_Laplace)
plt.title('Laplace Distribution')
plt.xlabel('x')
plt.ylabel('CDF')
plt.savefig('Laplace Distribution CDF')
```

```python
#location parameter  := 2 and scale parameter b := 2
mu = 2

def RiemannSum_Laplace_Variance(a, b, N): #Computing Left Riemann Sum
    delta_x = (b-a)/N 
    Variance = 0.0
    x = a     
    for i in range(N):
        Variance += (x-mu)*(x-mu)*Laplace(x,mu=2,b=2)*delta_x
        x = x + delta_x 
    return Variance
```

```python
var = RiemannSum_Laplace_Variance(-30, 30, 1000)
var
```

True value of variance = $2b^{2}$ = $2*2^{2}$ = 8

Computed approximate variance = 7.999

### Gumbel Distribution

```python
def Gumbel(x,mu,beta):
    z = (x - mu)/beta
    return ((1/beta)*exp(-z-exp(-z)))
```

```python
xlist1 = np.linspace(-15,15,num=1000)
ylist1 = [Gumbel(x,mu=1,beta=2) for x in xlist1] 

plt.figure(num=0,dpi=120)
plt.plot(xlist1,ylist1)
plt.title('Gumbel Distribution')
plt.xlabel('x')
plt.ylabel('PDF')
plt.savefig('Gumbel Distribution PDF')
```

```python
def RiemannSum_Gumbel(a, b, N): #Computing Left Riemann Sum
    cdf = []
    delta_x = (b-a)/N
    Csum = 0.0
    x = a     
    for i in range(N):
        Csum += Gumbel(x,mu=1,beta=2)*delta_x
        cdf.append(Csum)
        x = x + delta_x 
    return cdf

CDF_Gumbel = RiemannSum_Gumbel(-15,15,1000)

plt.figure(num=0,dpi=120)
plt.plot(xlist,CDF_Gumbel)
plt.title('Gumbel Distribution')
plt.xlabel('x')
plt.ylabel('CDF')
plt.savefig('Gumbel Distribution CDF')
```

```python
#location parameter  := 1 and scale parameter  := 2

def RiemannSum_Gumbell_Mean(a, b, N): #Computing Left Riemann Sum
    delta_x = (b-a)/N 
    Mean = 0.0
    x = a     
    for i in range(N):
        Mean += x*Gumbel(x,mu=1,beta=2)*delta_x
        x = x + delta_x 
    return Mean

mu = RiemannSum_Gumbell_Mean(-30,30,1000)
mu
```

```python
def RiemannSum_Gumbel_Variance(a, b, N): #Computing Left Riemann Sum
    delta_x = (b-a)/N 
    Variance = 0.0
    x = a     
    for i in range(N):
        Variance += (x-mu)*(x-mu)*Gumbel(x,mu=1,beta=2)*delta_x
        x = x + delta_x 
    return Variance

var = RiemannSum_Gumbel_Variance(-30, 30, 1000)
var
```

True value of variance = ${\displaystyle {\frac {\pi ^{2}}{6}}\beta ^{2}}$ = ${\displaystyle {\frac {\pi ^{2}}{6}}* 2 ^{2}}$ = 6.579

Computed approximate variance = 6.579

### Cauchy Distribution

```python
def Cauchy(x,x0,gamma):
    res = pi*gamma*(1+((x-x0)/gamma)**2)
    return (1/res)
```

```python
xlist2 = np.linspace(-15,15,num=1000)
ylist2 = [Cauchy(x,x0=0,gamma=1) for x in xlist2] 

plt.figure(num=0,dpi=120)
plt.plot(xlist2,ylist2)
plt.title('Cauchy Distribution')
plt.xlabel('x')
plt.ylabel('PDF')
plt.savefig('Cauchy Distribution PDF')
```

```python
def RiemannSum_Cauchy(a, b, N): #Computing Left Riemann Sum
    cdf = []
    delta_x = (b-a)/N
    Csum = 0.0
    x = a     
    for i in range(N):
        Csum += Cauchy(x,x0=0,gamma=1)*delta_x
        cdf.append(Csum)
        x = x + delta_x 
    return cdf

CDF_Cauchy = RiemannSum_Cauchy(-15,15,1000)

plt.figure(num=0,dpi=120)
plt.plot(xlist2,CDF_Cauchy)
plt.title('Cauchy Distribution')
plt.xlabel('x')
plt.ylabel('CDF')
plt.savefig('Cauchy Distribution CDF')
```
