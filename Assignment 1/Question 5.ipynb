```python
import numpy as np
import random
import matplotlib.pyplot as plt
np.random.seed(0)
```

```python
#Generating a dataset comprising a set of N real numbers drawn from the uniform distribution on [0, 1].
def generate_dataset_uniform(N):
    d = []
    for i in range(N):    
        val = np.random.uniform(0,1)
        d.append(val)  
    return d
```

Considering various dataset sizes N = 5, 10, 20, 40, 60, 80, 100, 500, 10^3, 10^4.
For each N,we need to repeat the following experiment M := 100 times:

(i) first, we need to generate the data

(ii) then, we need to compute the average µhat, and

(iii) finally,we need to measure the error between the computed average µhat and the true mean µ as |µhat − µtrue|.

Please find below calculation of true mean for Uniform[0,1].



```python
def experiment(N,M=100):
    error_list =[]
    µ_true=0.5
    for i in range(M):#Repeating the experiment M times
        dataset = generate_dataset_uniform(N) #Generating the data
        µ_hat = sum(dataset)/N        
        error = abs(µ_hat - µ_true)
        error_list.append(error)
    return error_list
```

```python
N = [5,10,20,40,60,80,100,500,1000,10000]
E_list = []
for n in N:
    E_list.append(experiment(n)) 
```

**Part I)For the uniform distribution, plot a single graph that shows the distribution of errors (across M repeats) for all values of N using a box-and-whisker plot. You may use the boxplot(.) function in Matlab.**

```python
plt.figure(dpi=200)
plt.boxplot(E_list,labels=['5','10','20','40','60','80','100','500','1000','10000'])
plt.xlabel('N')
plt.show()
plt.savefig('Box_plot_uniform.png')
```

**Part II: Repeat the above question by replacing the uniform distribution by a Gaussian distribution with µ := 0 
and σ^2:= 1.**

```python
def generate_dataset_normal(N):
    d = []
    for i in range(N):    
        val = np.random.normal()
        d.append(val)  
    return d

def experiment_normal(N,M=100):
    error_list =[]
    µ_true=0
    for i in range(M):#Repeating the experiment M times
        dataset = generate_dataset_normal(N) #Generating the data
        µ_hat = sum(dataset)/N        
        error = abs(µ_hat - µ_true)
        error_list.append(error)
    return error_list

E_list_normal = []
for n in N:
    E_list_normal.append(experiment_normal(n)) 
```

```python
plt.figure(dpi=200)
plt.boxplot(E_list_normal,labels=['5','10','20','40','60','80','100','500','1000','10000'])
plt.xlabel('N')
plt.show()
plt.savefig('Box_plot_normal.png')
```

**Part III) Interpret what you see in the graphs. What happens to the distribution of error as N increases?**

As N increases,the sample mean approaches the true mean of the distribution.Thus the error or the deviation of sample mean from true mean decreases as can be observed in the box plots. This is in conformance with **Law of Large Numbers**.
