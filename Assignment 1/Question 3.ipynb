**Question 3 : Simulate N := 10^4 independent random walkers (as discussed in class) along the real line, each walker starting at the origin, and each walker taking 10^3 steps each of length 10^âˆ’3.**

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
```

**Part I: Plot a histogram of the final locations of all the random walkers. You may use the hist(.) function in
Matlab.**

```python
def RandomWalk(N,d): #N is the number of steps, d is the step size
    # Each time depending on whether the binomial output is 0 or 1,the random walker takes a step of -d or +d respectively
    #If binomial output is 0, then step=(2*0 -1)*d=-d
    #If binomial output is 1, then step=(2*1 -1)*d=+d
    step = (2*np.random.binomial(1,.5,N)-1)*d
    walk = np.cumsum(step) #Adding up the steps
    return walk
```

```python
endpoints = []
for k in range(10**4):
    walk = RandomWalk(10**3,10**-3)
    endpoints.append(walk[-1]) #walk[-1] is the final location of an individual random walker

plt.figure(dpi=150)
plt.hist(endpoints,bins=200)
plt.title('Histogram of the final locations of all the random walkers')
plt.xlabel('Class Intervals')
plt.ylabel('Frequency')
plt.savefig('Histogram of the final locations of all the random walkers.png')
```

**Part II) For the first 10^3 walkers, plot space-time curves that show the path taken by each walker (as depicted
in the class slides). On the graph, draw each path in a different randomly-chosen color for better clarity.**

```python
import random
rand_colors = []
for j in range(10**3):
    rcolor = "#" + ''.join([random.choice('ABCDEF0123456789') for i in range(6)])
    rand_colors.append(rcolor)

plt.figure(num=0,dpi=150)

for k in range(10**3):
    particularWalk = RandomWalk(10**3,10**-3)
    #endpoints.append(particularWalk[-1])
    plt.plot(np.arange(10**3),particularWalk,rand_colors[k],linewidth=1)
plt.xlabel('Number of steps')
plt.ylabel('Position of walker')
plt.title('Space-time curves of random walkers')
plt.show()
plt.savefig('Space-time curves of random walkers.png')
```
