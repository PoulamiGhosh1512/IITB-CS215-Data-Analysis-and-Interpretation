**Part I) Write code to generate independent draws from PX(). Your code can use only the uniform random number generator rand() (no other generator). Submit this code.**

In this question,I have used the idea of Inverse Transform Sampling. 

If X is a continuous random variable with cdf $F_{X}$, then $F_{X}$ is distributed uniformly on [0,1]. Using this,I can simulate draws of X using draws from a uniform distribution and its inverse CDF function $F^{1}_{X}$.

I have generated a random sample using uniform random number generator and then used the inverse cdf of the distribution, in order to get a random sample of the distribution of interest.

Please find below computation of CDF and Inverse CDF for given distribution.



```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
np.random.seed(0)
```

```python
def inverse_cdf(y):# For M-shaped distribution
    if 0<=y<0.5:
        return -sqrt(1-2*y)
    else:
        return sqrt(2*y-1)
```

```python
def random_sample_generator(inv_cdf):#The inverse CDF function is passed as an argument to this function
    rn = np.random.rand() #Generates a random number between 0 and 1
    x = inv_cdf(rn) #Compute the corresponding x value
    return x
```

```python
random_sample_generator(inverse_cdf)
```

**Part II) Show plots of (i) the histogram (with 200 bins) and (ii) cumulative distribution function (CDF), both
using M := 10^5 draws from the PDF PX().**

```python
rng_list = []
for _ in range(10**5):
    rng_list.append(random_sample_generator(inverse_cdf))    
```

```python
plt.figure(num=0,dpi=200)
plt.hist(rng_list,bins=200)
plt.xlabel("x")
plt.ylabel("Frequency")
plt.title('Histogram of M-shaped distribution')
plt.savefig('histogram.png')
```

```python
freq, hbin = np.histogram(rng_list,bins=200)
```

```python
pdf = freq/10**5
plt.figure(dpi=200)
plt.plot(hbin[1:], pdf)
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('PDF')
```

```python
cdf = np.cumsum(pdf)
```

```python
plt.figure(dpi=200)
plt.plot(hbin[1:], cdf)
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('CDF')
plt.savefig('CDF.png')
```

**Part III) Use the code written in the previous sub-question to write code to generate independent draws from PYN(). Submit this code.**

```python
def random_sample_generator_Y(N):
    sum=0
    for _ in range(N):
        sum += random_sample_generator(inverse_cdf)
    return sum/N
```
